```python
data_frame = pd.read_csv("train.csv") 

# print(data_frame.head())
# print(data_frame.corr())

ID_column = data_frame.loc[:,['id']].values
data_frame.drop(columns = 'id', inplace= True)




# majority is 1.0

data_frame["CARTYPE_07_CAT"].fillna(data_frame['CARTYPE_07_CAT'].median(),inplace=True)
data_frame["REGISTRATION_03"].fillna(data_frame['REGISTRATION_03'].median(), inplace = True)
data_frame["CONDITION_05_CAT"].fillna(data_frame['CONDITION_05_CAT'].median(), inplace = True)
data_frame["CONDITION_02_CAT"].fillna(data_frame['CONDITION_02_CAT'].median(), inplace = True)
data_frame["CARTYPE_09_CAT"].fillna(data_frame['CARTYPE_09_CAT'].median(), inplace = True)
data_frame["CARTYPE_01_CAT"].fillna(data_frame['CARTYPE_01_CAT'].median(), inplace = True)
data_frame["CARTYPE_11"].fillna(data_frame['CARTYPE_11'].median(), inplace = True)
data_frame["CARTYPE_02_CAT"].fillna(data_frame['CARTYPE_02_CAT'].median(), inplace = True)
data_frame['CONDITION_04_CAT'].fillna(data_frame['CONDITION_04_CAT'].median(), inplace = True)

# continuous columns 
data_frame["CARTYPE_12"].fillna(data_frame['CARTYPE_12'].mean(), inplace = True)
data_frame["CARTYPE_14"].fillna(data_frame['CARTYPE_14'].median(), inplace = True)


# do either ways let's SEE
data_frame.drop(columns = 'CARTYPE_03_CAT', inplace= True)
data_frame.drop(columns = 'CARTYPE_05_CAT', inplace= True)
# creating other class for cartype_03_cat column 
# data_frame['CARTYPE_03_CAT'].fillna(-1, inplace = True)

# replacing all nans with -1 in cartype_05_cat
# data_frame['CARTYPE_05_CAT'].fillna(1, inplace = True)


# data_frame.isna().sum()


y = data_frame.loc[:,['target']]
data_frame.drop(columns = 'target', inplace= True)

X = data_frame.loc[:,:]

bnb = BernoulliNB()
bnb.fit(X, np.ravel(y))

yhat = bnb.predict(X)
print('Accuracy {:.2f}'.format(bnb.score(X, y)))

# TEST ALGORITHM
df2 = pd.read_csv("test.csv")
ID_column2 = df2.loc[:,['id']].values
df2.drop(columns = 'id', inplace = True)
df2.drop(axis="columns", labels="CARTYPE_03_CAT", inplace=True) 
df2.drop(axis="columns", labels="CARTYPE_05_CAT", inplace=True) 




X_test = df2.loc[:, :]
df2["CARTYPE_07_CAT"].fillna(df2['CARTYPE_07_CAT'].median(),inplace=True)
df2["REGISTRATION_03"].fillna(df2['REGISTRATION_03'].median(), inplace = True)
df2["CONDITION_05_CAT"].fillna(df2['CONDITION_05_CAT'].median(), inplace = True)
df2["CONDITION_02_CAT"].fillna(df2['CONDITION_02_CAT'].median(), inplace = True)
df2["CARTYPE_09_CAT"].fillna(df2['CARTYPE_09_CAT'].median(), inplace = True)
df2["CARTYPE_01_CAT"].fillna(df2['CARTYPE_01_CAT'].median(), inplace = True)
df2["CARTYPE_11"].fillna(df2['CARTYPE_11'].median(), inplace = True)
df2["CARTYPE_02_CAT"].fillna(df2['CARTYPE_02_CAT'].median(), inplace = True)
df2['CONDITION_04_CAT'].fillna(df2['CONDITION_04_CAT'].median(), inplace = True)

# continuous 
df2["CARTYPE_14"].fillna(df2['CARTYPE_14'].mean(), inplace = True)

# df2['CARTYPE_05_CAT'].fillna(1, inplace = True)

# df2.isna().sum()



Yhat_pred = bnb.predict(X_test)
print(Yhat_pred.max())
Yhat_pred = pd.DataFrame(Yhat_pred, columns = ['target'])
Yhat_pred["id"] = ID_column2
Yhat_pred = Yhat_pred[['id', 'target']]
# Yhat_pred = Yhat_pred[['target']]
Yhat_pred.to_csv("output_binomial.csv", index = False)


```
